# 2. Business Statement 
  To identify whether to give insurance claim to the person  who has claimed it.

## 2.1 Car Insurance Claim
Dataset Shape: 418 rows x 20 columns

1. ID: unique identifier
2. AGE: It is the age group of people
3. GENDER: Gender of the Applicant
4. RACE: If the person is in majority or minority
5. DRIVING_EXPERIENCE:Driving Experience of the person
6. EDUCATION: How much educated the person is
7. INCOME: To which class the person belong in terms of income
8. Credit_Score: Credit Score of the person
9. Vehicle_Owner: If the person owns a vehicle[1] or not[0]
10. Vehicle_Year: The vehicle bought before 2015
11. Married: If the person is married or not
12. Children: Does the person has children or not
13. Postal_Code: The postal code of the person
14. Annual Mileage: How much annually the person drives at what mileage
15. Vehicle_Type: Which type of vehicle that person owns
16. Speeding_Violations: How much speed violations are issused on the name of that person
17. DUIS: Dues on the person
18. Past_Accidents: How many times the person got into an accident
19. OUTCOMES: The target variable



Columns Types:
1. Continuous: 11
   - Explanatory Variables: 12 - float64 and int64
   - Target Variable: 1 - Float-64 
2. Categorical: 8

# Reading Data Set

```python
df = pd.read_csv('Car_Insurance_Claim.csv')
```

```python
df.isnull().sum()/df.shape[0]*100
```

```python
df.select_dtypes(include=(['int64', 'float64'])).isna().sum()
df_num_col = df.select_dtypes(include=(['int64', 'float64'])).columns
for c in df_num_col:
    df[c].fillna(df[c].median(), inplace=True)

df.select_dtypes(include=(['int64', 'float64'])).isna().sum()
```

```python
df.select_dtypes(include=('object')).isna().sum()
df_cat_col = df.select_dtypes(include=('object')).columns
for c in df_cat_col:
    df[c].fillna(df[c].mode().values[0], inplace=True)
    
df.select_dtypes(include=('object')).isna().sum()
```

```python
df.isnull().sum()
```

```python
X = df.drop('OUTCOME', axis=1)
y = df['OUTCOME']
```

```python
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(X, y, test_size = .3, random_state = 2022)
```

```python
# Break X_train in to 2 parts: X_train_cat, X_train_cont, same w/ X_test
X_train_cat = X_train.select_dtypes(exclude='number')
X_train_cat
```

```python
X_test_cat = X_test.select_dtypes(exclude='number')
X_test_cat
```

```python
X_train_cont = X_train.select_dtypes(include='number')
X_train_cont
```

```python
X_test_cont = X_test.select_dtypes(include='number')
X_test_cont
```

```python
from sklearn.preprocessing import OneHotEncoder

# Encoding on X_train_cat and X_test_cat
ohe = OneHotEncoder(drop = 'first')

X_train_cat_encoded = pd.DataFrame(ohe.fit_transform(X_train_cat).toarray(), columns = ohe.get_feature_names(list(X_train_cat.columns)))
X_test_cat_encoded = pd.DataFrame(ohe.transform(X_test_cat).toarray(), columns = ohe.get_feature_names(list(X_test_cat.columns)))
```

```python
from sklearn.preprocessing import StandardScaler

# Scaling on X_train_cont and X_test_cont
sc = StandardScaler()

X_train_cont_scaled = pd.DataFrame(sc.fit_transform(X_train_cont), columns = X_train_cont.columns)
X_test_cont_scaled = pd.DataFrame(sc.transform(X_test_cont), columns = X_test_cont.columns)
```

```python
X_train = pd.concat([X_train_cat_encoded, X_train_cont_scaled], axis = 1)
X_test = pd.concat([X_test_cat_encoded, X_test_cont_scaled], axis = 1)
```

```python
colors= ["YlGnBu", "Blues", "BuPu", "Greens", "PiYG"]
rnd.choice(colors)
```

# Metrics used: Accuracy Score, F-Beta Score, Classification Report, Confusion Matrix, ROCAUC Curve, Precision-Recall Curve

```python
from sklearn.metrics import accuracy_score, fbeta_score
from sklearn import metrics
from yellowbrick.contrib.wrapper import wrap
from yellowbrick.classifier import ClassificationReport, ConfusionMatrix, ROCAUC, PrecisionRecallCurve
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

#Created an obj for logistic regression 
log_reg = LogisticRegression(random_state = 10)
log_reg.fit(X_train, y_train)
```

```python
#Training Accuracy
log_reg.score(X_train, y_train)
```

```python
#Testing Accuracy
y_train_pred = log_reg.predict(X_train)
y_test_pred = log_reg.predict(X_test)
print(accuracy_score(y_test, y_test_pred))
print(accuracy_score(y_train, y_train_pred))
```

```python
precision_positive = metrics.precision_score(y_test, y_test_pred, pos_label=1)
precision_positive
```

```python
precision_positive = metrics.precision_score(y_train, y_train_pred, pos_label=1)
precision_positive
```

```python
recall_specificity = metrics.recall_score(y_test, y_test_pred, pos_label=1)
recall_specificity
```

```python
recall_specificity = metrics.recall_score(y_train, y_train_pred, pos_label=1)
recall_specificity
```

```python
fbeta_score(y_train, y_train_pred, beta = 1, average = None)
```

```python
fbeta_score(y_test, y_test_pred, beta = 1, average = None)
```

```python
visualizer = ClassificationReport(log_reg, support = True, color_bar = True, cmap = rnd.choice(colors))
visualizer.fit(X_train, y_train)        
visualizer.score(X_test, y_test)   
visualizer.show()
```

```python
visualizer = ConfusionMatrix(log_reg, support = True, color_bar = True, cmap = rnd.choice(colors))
visualizer.fit(X_train, y_train)        
visualizer.score(X_test, y_test)   
visualizer.show() 
```

```python
visualizer = ROCAUC(log_reg, support = True, color_bar = True, cmap = rnd.choice(colors))
visualizer.fit(X_train, y_train)        
visualizer.score(X_test, y_test)   
visualizer.show()
```
